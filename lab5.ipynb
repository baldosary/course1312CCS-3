### <span style="color:#5C6BC0">Looping</span>

- *Loops* allow repetitive execution of code blocks.

#### Two Main Types of Loops

- **for loop**: Iterates over sequences such as lists, tuples, and strings.
- **while loop**: Executes as long as a given condition remains `True`.

---

### 1.1 for Loop

- Uses the `for` keyword.
- Iterates over a sequence.
- Executes statements for each element in the sequence.

#### Syntax

```python
for variable in sequence:
    statements

````python
#Printing numbers from 1 to 5

for i in range(1, 6): 
    print(i)
````

### 1.2 while Loop

- Uses the `while` keyword.
- Repeats execution until a condition becomes `False`.
- Requires proper indentation and a valid condition.

#### Syntax

```python
while condition:
    statements

````python
#Printing numbers from 1 to 5 using while loop
i = 1
while i <= 5:
    print(i)
    i += 1
````

### 2. break Statement

- The `break` keyword stops a loop when a specific condition is met.
- Commonly used to exit `for` and `while` loops prematurely.

````python
#### Example: Stop loop when `i == 3`

for i in range(1, 6):
    if i == 3:
        break
    print(i)
````

### 3. continue Statement

- The `continue` keyword skips the current iteration of a loop.
- It immediately moves to the next iteration.
- Helps control program flow by avoiding unnecessary execution inside loops.

````python
#Skip printing 3
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
````

### 4. pass Statement

- The `pass` keyword acts as a placeholder for future code.
- Used when a statement is required syntactically but no action is needed.
- Commonly used in empty functions, loops, or conditional statements.

````python
#Placeholder inside a loop
for i in range(5):
    if i == 3:
         pass # Placeholder for future code
    else:
         print(i)
````
