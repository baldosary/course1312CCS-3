## <span style="color:#5C6BC0"> A) Variables </span>

**Variables** are containers for storing data values (named memory locations).

In Python:
- There is no command needed to declare a variable.
- A variable is created when you assign a value to it for the first time.


```python
# Example: Creating Variables
x = 5
y = "John"

print(x)
print(y)
```

## <span style="color:#5C6BC0"> B) Getting the Data Type (`type()`) </span>

You can get the data type of a variable using the `type()` function.


```python
x = 5
y = "John"

print(type(x))
print(type(y))
```

## <span style="color:#5C6BC0"> C) Casting (Typecasting) </span>

If you want to change the data type of a variable, you can do it using casting.
Common casting functions:
- `str()`
- `int()`
- `float()`


```python
x = str(3)    # '3'
y = int(3)    # 3
z = float(3)  # 3.0

print(x, type(x))
print(y, type(y))
print(z, type(z))
```

## <span style="color:#5C6BC0"> D) Operators </span>

Operators are used to perform operations on values and variables.
Python supports multiple types of operators, including:
- Arithmetic Operators
- Assignment Operators
- Relational/Comparison Operators
- Logical Operators
- Bitwise Operators
- Identity Operators
- Membership Operators


## <span style="color:#5C6BC0"> E) Arithmetic Operators </span>

Arithmetic operators are used for mathematical operations like:
- addition (+)
- subtraction (-)
- multiplication (*)
- division (/)
- exponent (**)
- modulus (%)
- floor division (//)


```python
print("1 + 4 =", 1 + 4)
print("10 - 5 =", 10 - 5)
print("5 * 2 =", 5 * 2)
print("10 / 2 =", 10 / 2)
print("2 ** 3 =", 2 ** 3)
print("15 % 4 =", 15 % 4)
print("17 // 5 =", 17 // 5)
```

## <span style="color:#5C6BC0"> Logical Operators </span>

Logical operators are used to combine conditional statements.

| Operator | Name | Description |
|--------|------|-------------|
| `and` | Logical AND | Returns True if **both** conditions are True |
| `or`  | Logical OR  | Returns True if **at least one** condition is True |
| `not` | Logical NOT | Reverses the result (True → False, False → True) |


```python
# Logical AND
print((8 > 9) and (2 < 9))
print((9 > 8) and (2 < 9))

# Logical OR
print((2 == 2) or (9 < 20))
print((3 != 3) or (9 > 20))

# Logical NOT
print(not (8 > 2))
print(not (2 > 10))
```

## <span style="color:#5C6BC0"> Bitwise Operators </span>

Bitwise operators work on **binary representations** of numbers and perform operations **bit by bit**.

Assume:
- `a = 10` → `1010`
- `b = 4`  → `0100`

| Operator | Name | Description |
|--------|------|-------------|
| `&` | AND | Sets each bit to 1 if both bits are 1 |
| `|` | OR | Sets each bit to 1 if one of the bits is 1 |
| `^` | XOR | Sets each bit to 1 if only one bit is 1 |
| `~` | NOT | Inverts all bits |
| `<<` | Left Shift | Shifts bits to the left |
| `>>` | Right Shift | Shifts bits to the right |


```python
a = 10
b = 4

print("a & b =", a & b)
print("a | b =", a | b)
print("~a =", ~a)
print("a ^ b =", a ^ b)
print("a >> 2 =", a >> 2)
print("a << 2 =", a << 2)
```
