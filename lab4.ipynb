## <span style="color:#5C6BC0"> Input and Output Functions </span>
- Input represents data given to the program.
- Output represents the result of the program.
- Input(p) is a predefined function.
- This function accepts input from the keyboard.
- This function takes a value from the keyboard and returns it as a string type.
- Based on the requirement we can convert from string to other types.

```python
weight=input ("Enter your weight")
print (weight, type(weight))
```

`print ()` is a function that prints a specific message to a standard output device.

```python
print ("Today is")
print("Thursday")
```

```python
print ('Today is'+"Thursday")
```

```python
print ("Today is", end =" ")
print("Thursday")
```

```python
print ('Today is',"Thursday")
```

```python
n1=10
n2=3
n3=1000
print (n1, n2, n3, sep ="->")
print(n1,"->", n2,"->",n3)
animal1="dog"
animal2="cat"
print("I adopted a {} and a {}.".format(animal1,animal2))
```

Converting from `string` to `int`:

- We can convert the string value into int value.
- int(p) is a predefined function
- This function converts to int data type.

```python
a="123"
print(a)
print(type(a))
b=int(a)
print(b)
print(type(b))
```

Converting from `float` to `int`:

- int(p) is a predefined function in Python.
- This function converts value into int data type.

```python
a=10000.45
print(a)
print(type(a))
b=int(a)
print(b)
print(type(b))
```
